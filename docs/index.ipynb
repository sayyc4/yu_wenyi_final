# 1. Housing Development Distribution Overview

Firstly, we plot the overview of the Housing Development by City Council District, this provides a general concepts about total housing units in different district, set up a base for further analysis. It can be seen that the total number distribution is generally uniform, so when we calculate the affordable housing development, we can directly use the amount without being affected by different total development amount.

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
housing_data = pd.read_csv('./geographicdatascience_python_finalproject/database/nychousingdb_23q4_csv/HousingDB_by_2024_CityCouncilDistrict.csv')
affordable_housing = pd.read_csv('./geographicdatascience_python_finalproject/database/Affordable_Housing_Production_by_Project_20241127.csv')
census_data = pd.read_csv('./geographicdatascience_python_finalproject/database/Community_Development_Block_Grant__CDBG__Eligibility_by_Census_Tract_-_CSV_20241221.csv')

# Update column names based on actual data
correct_district_column = 'councildst'  # City Council District
correct_housing_column = 'cenunits20'   # Housing Units (if this is the correct column)

# Summarize the housing development by council district
housing_summary = housing_data.groupby(correct_district_column)[correct_housing_column].sum().reset_index()

# Rename columns for clarity
housing_summary.rename(columns={correct_district_column: 'CityCouncilDistrict',
                                 correct_housing_column: 'TotalHousingUnits'}, inplace=True)

# Visualization: Housing development distribution
sns.barplot(data=housing_summary, x='CityCouncilDistrict', y='TotalHousingUnits')
plt.title('Housing Development by City Council District')
plt.xlabel('City Council District')
plt.ylabel('Total Housing Units')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
